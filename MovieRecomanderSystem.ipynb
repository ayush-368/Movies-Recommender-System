```python
import ast
```

```python
def extract_genres(txt):
    temp=[]
    txt = ast.literal_eval(txt)
    for i in txt:
        temp.append(i['name'])
        
    return temp
```

```python
df['genres'] = df['genres'].apply(extract_genres)
```

```python
df.head(1)
```

```python
df.iloc[1]['keywords']
```

```python
def extract_keywords(txt):
    temp=[]
    txt = ast.literal_eval(txt)
    for i in txt:
        temp.append(i['name'])
        
    return temp
```

```python
df['keywords'] = df['keywords'].apply(extract_genres)
```

```python
df.head(1)
```

```python
df.iloc[1]['cast']
```

```python
def extract_cast(txt):
    temp=[]
    counter=0
    txt = ast.literal_eval(txt)
    for i in txt:
        if(counter!=3):
            temp.append(i['name'])
            counter=counter+1
        else:
            break
    return temp
```

```python
df['cast'] = df['cast'].apply(extract_cast)
```

```python
df.iloc[1]['crew']
```

```python
def extract_director(txt):
    temp=[]
    txt = ast.literal_eval(txt)
    for i in txt:
        if(i['job']=='Director'):
            temp.append(i['name'])
            break
        
    return temp
```

```python
df['crew'] = df['crew'].apply(extract_director)
```

```python
df.head()
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","")for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df.head(1)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head(1)
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df = df.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#stemming
def stemming(txt):
    temp=[]
    for i in txt.split():
        i = ps.stem(i)
        temp.append(i)
        
    return " ".join(temp)
```

```python
df['tags'] = df['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities =cosine_similarity(vectors)
```

```python
similarities.shape
```

```python
sorted(list(enumerate(similarities[0])),reverse=True,key=lambda x:x[1])
```

```python
def recomdations(movie):
    ind = df[df['title']==movie].index[0]
    distances=similarities[ind]
    
    movies = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    movies_ind_list=[]
    for i in movies:
        movies_ind_list.append(i[0])
        
    for i in movies_ind_list:
        print(df.iloc[i]['title'])
```

```python
recomdations('Avatar')
```
